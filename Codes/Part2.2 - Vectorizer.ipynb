```python
import pandas as pd
```

```python
# input the labeled dataset
label_data0 = pd.read_csv('/content/drive/My Drive/CS5344/GP/label_data_string_df/part-00000-9fb4a9ee-be78-4663-8f3a-ac8775296021-c000.csv', header = None)
label_data1 = pd.read_csv('/content/drive/My Drive/CS5344/GP/label_data_string_df/part-00001-9fb4a9ee-be78-4663-8f3a-ac8775296021-c000.csv', header = None)
label_data2 = pd.read_csv('/content/drive/My Drive/CS5344/GP/label_data_string_df/part-00002-9fb4a9ee-be78-4663-8f3a-ac8775296021-c000.csv', header = None)
label_data3 = pd.read_csv('/content/drive/My Drive/CS5344/GP/label_data_string_df/part-00003-9fb4a9ee-be78-4663-8f3a-ac8775296021-c000.csv', header = None)
label_data = pd.concat([label_data0,label_data1,label_data2,label_data3],axis=0)
label_data.columns = ['Id', 'label', 'text']
print(label_data)
```

```python
# input the unlabeled dataset
main_data0 = pd.read_csv('/content/drive/My Drive/CS5344/GP/main_data_string_df/part-00000-db08faed-c3b7-4867-8399-c5eed41f7813-c000.csv', header = None)
main_data1 = pd.read_csv('/content/drive/My Drive/CS5344/GP/main_data_string_df/part-00001-db08faed-c3b7-4867-8399-c5eed41f7813-c000.csv', header = None)
main_data2 = pd.read_csv('/content/drive/My Drive/CS5344/GP/main_data_string_df/part-00002-db08faed-c3b7-4867-8399-c5eed41f7813-c000.csv', header = None)
main_data3 = pd.read_csv('/content/drive/My Drive/CS5344/GP/main_data_string_df/part-00003-db08faed-c3b7-4867-8399-c5eed41f7813-c000.csv', header = None)
main_data = pd.concat([main_data0,main_data1,main_data2,main_data3],axis=0)
main_data.columns = ['Id', 'text']
print(main_data)
```

```python
# input the evaluation dataset
eval_data0 = pd.read_csv('/content/drive/My Drive/CS5344/GP/eval_data_string_df/part-00000-d55c0795-2145-4786-9336-1a217d21f9c6-c000.csv', header = None)
eval_data1 = pd.read_csv('/content/drive/My Drive/CS5344/GP/eval_data_string_df/part-00001-d55c0795-2145-4786-9336-1a217d21f9c6-c000.csv', header = None)
eval_data2 = pd.read_csv('/content/drive/My Drive/CS5344/GP/eval_data_string_df/part-00002-d55c0795-2145-4786-9336-1a217d21f9c6-c000.csv', header = None)
eval_data3 = pd.read_csv('/content/drive/My Drive/CS5344/GP/eval_data_string_df/part-00003-d55c0795-2145-4786-9336-1a217d21f9c6-c000.csv', header = None)
eval_data = pd.concat([eval_data0,eval_data1,eval_data2,eval_data3],axis=0)
eval_data.columns = ['Id', 'label', 'text']
print(eval_data)
```

```python
# input the vocabulary
vocabulary = pd.read_csv('/content/drive/My Drive/CS5344/GP/vocabulary.csv', na_filter = False)
vocabulary = vocabulary['vocabulary'].values.tolist()
print(vocabulary)
```

```python
# CountVectorizer
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(vocabulary = vocabulary)
label_features = cv.fit_transform(label_data['text'])
main_features = cv.transform(main_data['text'])
eval_features = cv.transform(eval_data['text'])
```

```python
# TF-IDF Vectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
tf_idf = TfidfVectorizer(vocabulary = vocabulary)
label_features_tfidf = tf_idf.fit_transform(label_data['text'])
main_features_tfidf = tf_idf.transform(main_data['text'])
eval_features_tfidf = tf_idf.transform(eval_data['text'])
print(label_features_tfidf)
```
